```python
import matplotlib.pyplot as plt

def bar_chart(response):
    if not response or not isinstance(response, list):
        print("Invalid response format.")
        return

    # Use the first item in the response to dynamically determine keys
    sample_item = response[0]
    if not isinstance(sample_item, dict):
        print("Invalid response format.")
        return

    # Extract data for plotting
    x_field = list(sample_item.keys())[0]
    y_field = list(sample_item.keys())[1]

    x_data = [item[x_field] for item in response]
    y_data = [float(item[y_field]) for item in response]

    # Bar chart
    plt.bar(x_data, y_data)
    plt.xlabel(x_field.capitalize())
    plt.ylabel(y_field.capitalize())
    plt.title(f'Top {x_field.capitalize()}s by {y_field.capitalize()}')
    plt.xticks(rotation=45, ha="right")  # Rotate X-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent clipping of labels
    plt.show()

# Example usage
response = [
    {"name": "Road-150 Red, 56", "revenue": "1018375.642"},
    {"name": "Road-150 Red, 62", "revenue": "968995.516"},
    {"name": "Mountain-100 Black, 44", "revenue": "801020.1266"},
]

bar_chart(response)
```

```python
import re

def sanitize_sql(sql_query):
    """
    Sanitize SQL query by replacing dots in aliases with underscores.
    """
    return re.sub(r'(\b[a-zA-Z_]\w*\b)\.', lambda match: f"{match.group(1)}_", sql_query)

# Example usage:
sql_query = """
SELECT
    hr.d."Name" as department_name,
    COUNT(CASE WHEN hr.e.gender = 'M' THEN 1 END) :: FLOAT / GREATEST(COUNT(CASE WHEN hr.e.gender = 'F' THEN 1 END), 1) AS male_to_female_ratio
FROM
    humanresources.department as hr.d
JOIN
    hr.d.departmentid = hr.edh.departmentid
JOIN
    humanresources.employee as hr.e
ON
    hr.edh.businessentityid = hr.e.businessentityid
GROUP BY
    department_name
ORDER BY
    department_name ASC;
"""

sanitized_query = sanitize_sql(sql_query)
print(sanitized_query)
```

```python
import os

def print_directory_structure(folder_path, indent=''):
    print(indent + f'[{os.path.basename(folder_path)}]')
    indent += '    '

    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        if os.path.isdir(item_path):
            print_directory_structure(item_path, indent)
        else:
            print(indent + f'{item}')

# Using double backslashes
project_folder_path = 'D:\\Programming\\Projects\\Test_NQL\\NQL_AI_Agent'

# or using a raw string
# project_folder_path = r'D:\Programming\Projects\Test_NQL\NQL_AI_Agent'

print_directory_structure(project_folder_path)
```

```python
def merge_sort(input_string):
    """
    Sorts a given string using merge sort algorithm, case insensitive.

    Args:
        input_string (str): The string to be sorted.

    Returns:
        str: The sorted string.
    """
    if len(input_string) <=1:
        return input_string
    
    mid = len(input_string) // 2
    left_half = merge_sort(input_string[:mid].lower()) # convert to lowercase
    right_half = merge_sort(input_string[mid:].lower()) # convert to lowercase

    return merge(left_half, right_half)

def merge(left, right):
    """
    Merges two sorted lists of characters.

    Args:
        left (list): A sorted list of characters.
        right (list): A sorted list of characters.

    Returns:
        list: The merged and sorted list of characters.
    """
    result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]: # use "<=" to handle equal elements
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result += left[i:]
    result += right[j:]

    return ''.join(result)
```
